# –õ–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ. –û—Å–Ω–æ–≤–Ω—ã–µ –ø–æ–Ω—è—Ç–∏—è. –ß–∞—Å—Ç—å 2

**1.** –ù–∞–π–¥–∏—Ç–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ ùë•,ùë¶ $\in \mathbb{‚Ñù}$ : –∞) ùë•=(0,‚àí3,6), ùë¶=(‚àí4,7,9); –±) ùë•=(7,‚àí4,0,1), ùë¶=(‚àí3,1,11,2).

**–†–µ—à–µ–Ω–∏–µ**

–í –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –±–∞–∑–∏—Å–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Å—É–º–º–æ–π –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–π —Å–æ–æ—Ç–≤–µ—Ç—Å–≤—É—é—â–∏—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç:

a)
$\overrightarrow{x}\cdot{}\overrightarrow{y}=x_1y_1 + x_2y_2+x_3y_3=0-21+54 =33$

–±)
$\overrightarrow{x}\cdot{}\overrightarrow{y}=x_1y_1 + x_2y_2+x_3y_3+x_4y_4=-21-4+0+2 =-23$

–Ω–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–∞ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞ python

```python
import numpy as np
print(f'–¥–ª—è –≤–∞—Ä–∏–∞–Ω—Ç–∞ –∞) —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ => {np.array([0,-3,6]).dot(np.array([-4,7,9]))}')
print(f'–¥–ª—è –≤–∞—Ä–∏–∞–Ω—Ç–∞ –±) —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ => {np.array([7,-4,0,1]).dot(np.array([-3,1,11,2]))}')
```

**2.**  –ù–∞–π–¥–∏—Ç–µ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤  x(4,2,4)  –∏  y(12,3,4)  –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.

**–†–µ—à–µ–Ω–∏–µ**

–î–ª—è –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Ç—Å—Ç–≤–∞ $\mathbb{R}^3$ –µ–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞   $\overrightarrow{x}$–æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –≤—ã—Ä–∞–∂–µ–Ω–∏–µ–º: $\|\overrightarrow{x}\|=\sqrt{x_1^2+x_2^2+x_3^2}$. <br>
–¢–æ–≥–¥–∞ –µ–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –ø–µ—Ä–≤–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ —Ä–∞–≤–Ω–∞  $\|\overrightarrow{x(x_1,x_2, x_3)}\|=\sqrt{16+4+16}=6$ <br>
–ï–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤—Ç–æ—Ä–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ —Ä–∞–≤–Ω–∞ $\|\overrightarrow{y(y_1,y_2, y_3)}\|=\sqrt{144+9+16}=\sqrt{169}=13$ 

—É–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ —Ä–∞–≤–µ–Ω x –∏ —Ä–∞–≤–µ–Ω <br>
$\varphi = acos(\frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|})=acos(\frac{4*12+2*3+4*4}{6*13})\approx{26.2}^0$.
–ù–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω –∫–æ–¥ —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –Ω–∞ python 

```python
import math
from numpy.linalg import norm
x = np.array([4,2,4])
y = np.array([12,3,4]) 
norm_x = norm(x)
norm_y = norm(y)
print(f'–ù–æ—Ä–º–∞ –ø–µ—Ä–≤–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ —Ä–∞–≤–Ω–∞ => {norm_x}')
print(f'–ù–æ—Ä–º–∞ –≤—Ç–æ—Ä–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ —Ä–∞–≤–Ω–∞ => {norm_y}')
print(f'—É–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ —Ä–∞–≤–µ–Ω => {math.degrees(math.acos(x.dot(y)/(norm_x*norm_y)))} ')
x.dot(y)/(norm_x*norm_y)
```

**–†–µ—à–µ–Ω–∏–µ**


–õ–∏–Ω–µ–π–Ω–æ –µ–≤–∫–ª–∏–¥–æ–≤–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –¥–æ–ª–∂–Ω–æ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—Ç—å 4 –∞–∫—Å–∏–æ–º–∞–º:

–ü—Ä–∏ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫–∞–∂–¥–æ–π –∏–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ü–∏–π —Å–∫–∞–ª—è—Ä–Ω–æ–≥–æ —É–º–Ω–æ–∂–µ–Ω–∏—è –Ω–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–∏—è 4 –∞–∫—Å–∏–æ–º–∞–º, –≤–≤–µ–¥–µ–º —Ç—Ä–∏ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞:
