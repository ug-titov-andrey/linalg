# Линейные преобразования


**1.** Найдите собственные векторы и собственные значения для линейного оператора, заданного матрицей:

$A=\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix}.$

**Решение**

Найдем собственные значения линейного оператора:

$\begin{vmatrix}
-1 - \lambda & -6\\ 
2 & 6 - \lambda
\end{vmatrix}$ = 0
$(-1-\lambda)(6-\lambda)+12=0$

$\lambda^2-5 \lambda +12=0$

$\lambda_1=2, \lambda_2=3 $

$\lambda_1=2$  <br>
$\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} =2\begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} $

$\begin{cases}
-x_1 - 6x_2=2x_1\\
2x_1+6x_2=2x_2
\end{cases}
\begin{cases}
-3x_1 - 6x_2=0\\
2x_1+4x_2=0
\end{cases}
\begin{cases}
x_1=-2x_2\\
6x_2 - 6x_2=0
\end{cases} 
$ => 

$x_2$  может быть любым. Примем, например,  $x_2 =1$ => $x_1=-2$

Собственные вектора для собственного значения  2 находятся на  прямой $x_1=-2x_2$ 

$\lambda_1=3$  <br>
$\begin{pmatrix}
-1 & -6\\ 
2 & 6
\end{pmatrix} \begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} =3\begin{pmatrix}
x_1\\ 
x_2
\end{pmatrix} $

$\begin{cases}
-x_1 - 6x_2=3x_1\\
2x_1+6x_2=3x_2
\end{cases}
\begin{cases}
-4x_1 - 6x_2=0\\
2x_1+3x_2=0
\end{cases}
\begin{cases}
-2x_1-3x_2=0\\
2x_1 + 3x_2=0
\end{cases} 
$ => 

$x_1 = -\frac{3}{2}x_2$. <br> $x_2$  может быть любым числом. Примем, например $x_2 =1$ => $x_1= -\frac{3}{2}$ 
Собственные вектора для собственного значения  3 находятся на  прямой $x_1= -\frac{3}{2}x_2$ 

**2.** Дан оператор поворота на 180 градусов, задаваемый матрицей:

$\textbf{A}=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$

Покажите, что __любой__ вектор считается для него собственным.

**Решение**

Матрица преобразования является диагональной, со значением -1 на главной диагонали. Диагональная матрица осуществляет растяжение. Раскрывая  $\textbf{A}x=\lambda x$, можно получить выражение 

$\begin{cases}
x_1= -\lambda x_1 \\
x_2= -\lambda x_2
\end{cases} 
$, имеющее решение при $\lambda=-1$. При этом получается тождественное преобразование, перевододящее каждый вектор пространства $(x_1, x_2)$ в себя 

**3.** Пусть линейный оператор задан матрицей:

$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$

Установите, считается ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

**Решение**

Составим характеристическое уравнение: 
$\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}\begin{pmatrix}
1\\ 
1
\end{pmatrix}   = \lambda \begin{pmatrix}
1\\ 
1
\end{pmatrix}$  

Раскрывая это уравнение, получим
$\begin{cases}
2= \lambda \cdot 1 \\
2= \lambda \cdot 1
\end{cases} $ => вектор $x=(1,1)$ является собственным вектором линейного оператора $\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$

__4.__ Пусть линейный оператор задан матрицей:

$$A=\begin{pmatrix} 0 & 3 & 0\\ 3 & 0 & 0\\ 0 & 0 & 3 \end{pmatrix}.$$

Установите, считается ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

**Решение**

Составим характеристическое уравнение: 
$\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0 \\
0 & 0 & 3 
\end{pmatrix}\begin{pmatrix}
3\\ 
-3\\
-4
\end{pmatrix}   = \lambda \begin{pmatrix}
3\\ 
-3\\
-4
\end{pmatrix}$  

Раскрывая это уравнение, получим
$\begin{cases}
-3= \lambda \\
3= -\lambda \\
3= \lambda
\end{cases} $. Система не имеет решения  => вектор  $x=(3, -3, -4)$ не является собственным вектором линейного оператора$A=\begin{pmatrix} 0 & 3 & 0\\ 3 & 0 & 0\\ 0 & 0 & 3 \end{pmatrix}.$
