**5.** –ù–∞–ø–∏—à–∏—Ç–µ –Ω–∞ Python —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–∏—è –¥–≤—É—Ö –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü, –Ω–µ –∏—Å–ø–æ–ª—å–∑—É—è NumPy.




```python
def matrix_mult(a,b):
    assert (len(a[0]) == len(b)), '–ß–∏—Å–ª–æ —Å—Ç–æ–ª–±—Ü–æ–≤ –ø–µ—Ä–≤–æ–π –º–∞—Ç—Ä–∏—Ü—ã –¥–æ–ª–∂–Ω–æ —Å–æ–≤–ø–∞–¥–∞—Ç—å —Å —á–∏—Å–ª–æ–º —Å—Ç—Ä–æ–∫ –≤—Ç–æ—Ä–æ–π –º–∞—Ç—Ä–∏—Ü—ã'
    out = []
    for y in range(len(a)):
        temporows = []
        for x in range(len(b[0])):
            tempo = 0 
            for ind in range(len(a[0])):
                tempo+=a[y][ind]*b[ind][x]
                #print(a[y)
            temporows.append(tempo)
        out.append(temporows)
    return out  
    
A= [[1,3,5,-5], [10,15,50,0], [2,1,19,-120]]
B= [[9,15,7], [1,18,52], [19,10,5], [-100,321,519]]
print(f'–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü =>{matrix_mult(A,B)}')

print('–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å –ø–æ–º–æ—â—å—é numpy')
import numpy as np
print(np.dot(A,B))    
```
